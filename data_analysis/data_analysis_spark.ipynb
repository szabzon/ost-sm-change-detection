```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

## Data Loading

```python
# Create a spark session
spark = SparkSession.builder.appName("HAI-Preprocessing").getOrCreate()

# Load dataset
hai_data = spark.read.format("csv").option("header", "true").load("/Users/emmatosato/Documents/UNI Locale/Erasmus/Draft OST/end-train1.txt")
```

## Data set insights

## Renaming columns

```python
tempList = [] 
for col in hai_data.columns:
    new_name = col.strip()
    new_name = "".join(new_name.split())
    new_name = new_name.replace('.','')
    tempList.append(new_name) 
print(tempList) 

hai_data = hai_data.toDF(*tempList) 
```
